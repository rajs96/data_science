# Running all the models

Here, we will compile all the models we ran into one notebook. We will also convert to using RandomizedSearchCV for all the models, so we can test over larger hyperparameter spaces.

Let's import some modules we may need later.

```python
# importing some needed modules
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# importing the data
filename = 'drugAbuse_data_cleaned_model_ready.csv'
df = pd.read_csv(filename)
df.drop(columns=['Unnamed: 0'],inplace=True)

# more modules 
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve

# function to describe model results
# model must already be fitted
def model_results(model,X,y):
    
    y_pred = model.predict(X)
    print("Results")
    print("\n")
    # accuracy score
    print("Accuracy Score: {}".format(accuracy_score(y_pred,y)))

    # classification report 
    print("Classification Report: \n {}".format(classification_report(y,y_pred)))

    # confusion matrix 
    print("Confusion Matrix: \n {}".format(confusion_matrix(y,y_pred)))

    # AUC score 
    print("AUC Score: {}".format(roc_auc_score(y,model.predict_proba(X)[:,1])))

    # Generate ROC curve values: fpr, tpr, thresholds
    fpr, tpr, thresholds = roc_curve(y, model.predict_proba(X)[:,1])

    # Plot ROC curve
    plt.plot([0, 1], [0, 1], 'k--')
    plt.plot(fpr, tpr)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.show()
```

# Logistic Regression Model

We will implement logistic regression as we did in the logistic regression notebook. We will try using both undersampling and balanced class weights.

# Training Model Results - Logistic Regression with Undersampling

```python
%%time
# import pipeline, scaler, and logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV
from imblearn.pipeline import Pipeline as pl
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import RandomUnderSampler
from sklearn.preprocessing import StandardScaler

# create our X and y
X = df.drop(columns=['Severe Psychological Distress Past Month']).values
y = df['Severe Psychological Distress Past Month']

# no class weights
steps = [('scaler',StandardScaler()),
        ('rus',RandomUnderSampler(random_state=42)),
        ('logistic',LogisticRegression(penalty='l2'))]

pipeline = pl(steps)

c_space = np.logspace(-5, 8, 15)
parameters = {'logistic__C': c_space}

# Create train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=24)

# Instantiate the GridSearchCV object: cv
cv = RandomizedSearchCV(pipeline, parameters,cv=5,scoring='f1')

cv.fit(X_train, y_train)

# Print the tuned parameters and score
print("Tuned Logistic Regression Parameters: {}".format(cv.best_params_)) 
# see results on training data 
model_results(cv,X_train,y_train)
```

# Testing Model Results - Logistic Regression with Undersampling

```python
model_results(cv,X_test,y_test)
```

# Training Model Results - Logistic Regression with Balanced Class Weights

```python
# balanced class weights
steps = [('scaler',StandardScaler()),
        ('logistic',LogisticRegression(penalty='l2',class_weight='balanced'))]

pipeline = pl(steps)

c_space = np.logspace(-5, 8, 15)
parameters = {'logistic__C': c_space}

# Create train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=27)

# Instantiate the GridSearchCV object: cv
cv = RandomizedSearchCV(pipeline, parameters,cv=5,scoring='f1')

cv.fit(X_train, y_train)

# Print the tuned parameters and score
print("Tuned Logistic Regression Parameters: {}".format(cv.best_params_)) 
# see results on training data 
model_results(cv,X_train,y_train)
```

# Testing Model Results - Logistic Regression with Balanced Class Weights

```python
model_results(cv,X_test,y_test)
```

# Random Forest Model

We will implement a random forest model as we did in the random forest notebook. As with logistic regression, we will try using undersampling and balanced class weights.

# Training Model Results - Random Forest with Undersampling

```python
%%time
from sklearn.ensemble import RandomForestClassifier
# no need to scale with a random forest
steps = [('rus',RandomUnderSampler(random_state=22)),
         ('random_forest',RandomForestClassifier())]

pipeline = pl(steps)
          
n_trees = [100,250,500,1000,1500,2000]
possible_max_depths = [1,2,3,4,5,None]
# tuning some more hyperparameters this time
param_grid = {'random_forest__n_estimators':n_trees,
             'random_forest__max_depth':possible_max_depths,
             'random_forest__max_features':['log2','sqrt'],
              'random_forest__criterion':['gini','entropy']}
              

cv = RandomizedSearchCV(pipeline,param_grid,cv=5,scoring='average_precision')


# Create train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=28)

# fit the model
cv.fit(X_train, y_train)

# Print the tuned parameters and score
print("Tuned Random Forest Parameters: {}".format(cv.best_params_)) 
print("Best score on training set is {}".format(cv.best_score_))

model_results(cv,X_train,y_train)
```

# Testing Model Results - Random Forest with Undersampling

```python
model_results(cv,X_test,y_test)
```

# Training Model Results - Random Forest with Balanced Class Weights

```python
%%time

random_forest = RandomForestClassifier(class_weight='balanced')
param_grid = {'n_estimators':n_trees,
             'max_depth':possible_max_depths,
             'max_features':['log2','sqrt'],
              'criterion':['gini','entropy']}

cv = RandomizedSearchCV(random_forest,param_grid,cv=5,scoring='f1')

# Create train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=23)

# fit the model
cv.fit(X_train, y_train)

# Print the tuned parameters and score
print("Tuned Random Forest Parameters: {}".format(cv.best_params_)) 

model_results(cv,X_train,y_train)
```

# Testing Model Results - Random Forest with Balanced Class Weights

```python
model_results(cv,X_test,y_test)
```

# Boosting

Now, we will try 3 different boosting approaches: AdaBoost, Gradient Boosting, and XGBoost. We will try balanced trees and random undersampling for our AdaBoost Classifier, and just random undersampling with our Gradient Booster and XGBooster.

# AdaBoost with Random Under-sampling - Training Results

```python
%%time
from sklearn.ensemble import AdaBoostClassifier

steps = [('scaler',StandardScaler()),
        ('rus',RandomUnderSampler(random_state=35)),
        ('adaboost',AdaBoostClassifier())]

pipeline = pl(steps)

n_trees = [100,250,500,1000,1500,2000]
learning_rates = [10,1,0.1,0.01,0.001]

parameters = {'adaboost__n_estimators':n_trees,
             'adaboost__learning_rate':learning_rates}

cv = RandomizedSearchCV(pipeline, parameters, cv=5,scoring='f1')

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=36)

cv.fit(X_train,y_train)

print("Tuned AdaBoost Parameters: {}".format(cv.best_params_)) 
model_results(cv,X_train,y_train)
```

# AdaBoost with Random Undersampling - Test Results

```python
model_results(cv,X_test,y_test)
```

# AdaBoost with Balanced Class-weight Decision Trees and Random-Undersampling - Training Results

```python
%%time
from sklearn.tree import DecisionTreeClassifier

steps = [('scaler',StandardScaler()),
         ('rus',RandomUnderSampler(random_state=20)),
        ('adaboost',AdaBoostClassifier(base_estimator=DecisionTreeClassifier(class_weight='balanced')))]

pipeline = pl(steps)

n_trees = [100,250,500,1000,1500,2000]
learning_rates = [10,1,0.1,0.01,0.001]

parameters = {'adaboost__n_estimators':n_trees,
             'adaboost__learning_rate':learning_rates}

cv = RandomizedSearchCV(pipeline, parameters, cv=5,scoring='f1')

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=39)

cv.fit(X_train,y_train)

print("Tuned AdaBoost Parameters: {}".format(cv.best_params_)) 
model_results(cv,X_train,y_train)
```

```python
model_results(cv,X_test,y_test)
```

# Gradient Boosting with Random Under-sampling - Training Results

```python
%%time 
from sklearn.ensemble import GradientBoostingClassifier

steps = [('scaler',StandardScaler()),
        ('rus',RandomUnderSampler(random_state=35)),
        ('gradient',GradientBoostingClassifier(random_state=40))]

pipeline = pl(steps)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=29)

n_trees = [100,250,500,1000,1500,2000]
learning_rates = [10,1,0.1,0.01,0.001]

parameters = {'gradient__n_estimators':n_trees,
             'gradient__learning_rate':learning_rates}

cv = RandomizedSearchCV(pipeline, parameters,cv=5,scoring='average_precision')

cv.fit(X_train, y_train)

print("Tuned Gradient Boosting Parameters: {}".format(cv.best_params_)) 
model_results(cv,X_train,y_train)
```

# Gradient Boosting with Random Under-sampling - Testing Results

```python
model_results(cv,X_test,y_test)
```
